### 1. Field of Complex Numbers

Consider an equation $x^2 = -1$. If we just use the field of real numbers, we could not find a solution to this equation.
To overcome this problem, __i__ was invented, defined as _square root of minus 1_ ($\sqrt{-1}).$
$$=> i^2 = -1$$

* Using __i__ we can find solution of any equation of the form $x^2\ =\ -b$ as x = $\sqrt{b}\ times\ i$.
* The product of a real nnumber and __i__ is called an _imaginary number_.

Python supports complex numbers. j plays the role of i. But i the imaginary number is written as 1j and not just j to avoid confusion with variable j.

```python
# Eg. 
3j
```

__Overloading in python__

In programming languages, use of the same name for different procedures operating on values of different datatypes is called _overloading_.

```python
# For e.g. let's define a procedure
def solve1(a,b,c): return (c-b)/a
# we will use it to solve 10x + 5 = 30
solve1(10,5,30)
```

```python
# But we can also use this procedure to solve equations involving complex numbers.
solve1(10+5j,5,20)
```

This is possible because the correctness of the procedure depends only on if the operations that we are doing are defined on that number or not. The field of the numbers does not matter.

#### Translation

If we add 1+2i to out set S we defined earlier, we are shifting the image 1 unit to the right and 2 units up.
This transformation of numbers in S is called _translation_.

Using Translation, we can move a picture anywhere in the complex plane.

__Problem 1.4.5__
Given 2 complex numbers $z_1$ and $z_2$, we can translate $z_1$ to $z_2$ using the translation
$$f(z_1) = z_1 + (z_2 -z_1)$$
Similarly, we can use a transaltion, to move $z_2$ to $z_1$.

__Complex numbers as arrows__

We can represent transaltion f(z) as arrows. The tail of the arrow is at complex number z and the head of the arrow is at located at f(z).

```python
# Translation as arrows
Image(filename = "/Users/atyagi/Desktop/Coding the Matrix - Python/Coding-the-matrix-Code/Images/translation.png",
      width = 500, height =500)
```

__Composing translations__

If $f_1(z)$ = $z_1$ + z and $f_2(z)$ = $z_2$ +z are two translations, then their composition is also a translation.

$$(f_2\ o\ f_1)\ =\ f_2(f_1(z))$$
$$=\ f_2(z_1+z)$$
$$=\ z_1\ +\ z_2\ +\ z$$

__Multiplying a complex number with a positive real number__

Multiplying a set of complex numbers with a number between 0 and 1 will shrink the image i.e. it will bring the points closer, this operation is called scaling. Similarly, multiplying with a number greater than 1 will move points farther away.

```python
# Task 1.4.7
# Scaling
plot({1/2 * z for z in S},4)
```

__Multiplying complex numbers by a negetive number__

Multiplying a set of complex number by -1 will rotate the image by 180 degrees.

```python
# Rotation of a plot 
S1 = np.array([2 + 2j,3 + 2j,1.75 + 1j,2 + 1j,2.25 + 1j,2.5+1j,2.75+1j,3+1j,3.25+1j])
plt.scatter(S1.real,S1.imag,color = 'b') # image from S1
plt.scatter(S1.real*(-1),S1.imag*(-1),color = 'r') #image after multiplying with -1.
plt.axhline(y = 0, color = 'k', linestyle = '--')
plt.axvline(x = 0, color = 'k', linestyle = '--')
plt.xticks(range(-5,6,1))
plt.yticks(range(-5,6,1))
plt.show()
# Blue color is the original image and red color is the image after multiplying with -1.
# We can see that the image is rotated 180 degrees.
```

__Multiplying complex numbers by i__

Multiplying a set of complex number with i will rotate the image by 90 degrees. As the coordinates now become (-y,x) from (x,y).

```python
S1 = np.array([2 + 2j,3 + 2j,1.75 + 1j,2 + 1j,2.25 + 1j,2.5+1j,2.75+1j,3+1j,3.25+1j])
S2 = S1 * (1j) # Multiply the number with i.
plt.scatter(S1.real,S1.imag,color = 'b') # image from S1
plt.scatter(S2.real,S2.imag,color = 'r') #image after multiplying with i.
plt.axhline(y = 0, color = 'k', linestyle = '--')
plt.axvline(x = 0, color = 'k', linestyle = '--')
plt.xticks(range(-5,6,1))
plt.yticks(range(-5,6,1))
plt.show()
# Blue color is the original image and red color is the image after multiplying with i.
# We can see that the image is rotated 90 degrees.
```
